```python
import pandas as pd
```

```python
#Import File
data_file = "worldcitiespop.csv"
```

```python
#Read CSV
data_df = pd.read_csv(data_file)
data_df.head()
```

```python
#Drop Nulls
dropnulls = data_df.dropna()
dropnulls.head()
```

```python
#Dropped City Column
clean_df = dropnulls.drop('City', axis=1)
clean_df.head()
```

```python
#Renamed AccentCity to City
rename_AccentCity = clean_df.rename(columns={'AccentCity' : 'City'})
rename_AccentCity.head()
```

```python
sorted_list = rename_AccentCity.sort_values(by='Population', ascending=False)
sorted_list.head()
```

```python
top100 = sorted_list[:100]
```

```python
reset_index = top100.reset_index(drop=True)
reset_index.head()
```

```python
reset_index.to_csv('Cleaned_Top_100_Cities')
```
